```python
## Test of combined pressure & tension kernel
## NB need net repulsion to prevent particles collapsing into a super dense mass. 
## Hence the surface tension coefficent must be small enough, that the radial integral over the smoothing basis is strongly positive.
sr,r = symbols('sr,r')
sr=1
# NB gives equi-pressure radius = 0.5.
CPkern = 1.2*(sr - r )**3 - (0.4)*(sr - r )**2
plt1 = plot(CPkern,(r,0,sr))
```

```python
plt2 = plot(integrate(4*pi*r**2*CPkern,r),(r,0,sr))
```

```python
integrate(4*pi*r**2*CPkern,(r,0,sr))
```

```python
plt2 = plot(integrate(4*pi*r**2*HypKern*CPkern, r), (r,0.0,1))
```
